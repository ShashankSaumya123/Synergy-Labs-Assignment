```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.transform import rescale
from skimage.filters import try_all_threshold
from skimage.filters import threshold_minimum, threshold_otsu, threshold_isodata, threshold_mean
from skimage.exposure import equalize_adapthist, equalize_hist, rescale_intensity
from skimage.morphology import reconstruction
from skimage.morphology import erosion, dilation, opening, closing
```

```python
img = imread('D:\Application Assignments\Synergy-Labs-Assignment\photo.jpg')
img = rescale(img, 0.4, anti_aliasing=False)
img1 = equalize_adapthist(img)
img2 = equalize_hist(img)
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(8, 2))

ax0.imshow(img, cmap = 'gray')
ax0.set_title("Original image")
ax0.axis('off')
ax1.imshow(img1, cmap='gray')
ax1.set_title("CLAHE")
ax1.axis('off')
ax2.imshow(img2, cmap='gray')
ax2.set_title("AHE")
ax2.axis('off')

fig.tight_layout()
```
